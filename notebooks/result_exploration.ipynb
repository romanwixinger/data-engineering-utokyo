```python
class ParameterRecorder(Recorder): 
    
    def _harmonize_time(self):
        self._table_df["datetime"] = pd.to_datetime(self._table_df["Time"])
        self._table_df["datetime_μs"] = self._table_df["datetime"].apply(lambda s: str(s) +".000000")
        self._table_df["datetime_ms"] = self._table_df["datetime"].apply(lambda s: str(s)+".000")
        self._table_df["timestamp"] = self._table_df["datetime"].values.astype(np.int64)
```

```python
parameter_recorder = ParameterRecorder(filepath=parameter_path, has_metadata=False, delimiter=";", encoding="utf-8")
```

```python
class QueryHelper(object): 
    
    @classmethod
    def query_valid_datetime(cls, df: pd.DataFrame) -> pd.DataFrame: 
        return df[~df["datetime"].isnull()]["datetime"]
    
    @classmethod
    def query_column_not_nan(cls, df: pd.DataFrame, columns: str or list) -> pd.DataFrame: 
        return df[df[columns].notna()]
```

```python
class CycleRecorder(object):
    """ Special recorder which tracks all results. Is not based on Recorder 
        like the other recorders, but implements the same methods.
    """
    
    def __init__(self, image_recorder, ssd_recorder, parameter_recorder):
        # Recorders
        self.parameter_recorder = parameter_recorder # Frequency parameter set in each cycle
        self.image_recorder = image_recorder         # Result of the analysis of the camera images
        self.ssd_recorder = ssd_recorder             # Result of the analysis of the peaks in the SSD data
        
        # Caching
        self.p_df = self.parameter_recorder.get_table().sort_values(by=['timestamp'])
        self.p_series = pd.Series(self.p_df["timestamp"])
        
    def get_table(self):
        # Load
        image_df = self._load_and_clean_image_df()
        ssd_df = self._load_and_clean_ssd_df()
        self._update_parameter_df()
        
        # Join image and ssd
        result_df = pd.concat([image_df, ssd_df], ignore_index=True)
        
        # Add parameter data
        df = self._add_parameter_data(result_df)
        
        return df
        
    def get_data(self):
        raise Exception("Method get_data() in CycleRecorder is not implemented.")
    
    def get_metadata(self):
        raise Exception("Method get_metadata() in CycleRecorder is not implemented.")
    
    def _closest_parameter_row(self, timestamp: int, threshold: int=360 * 24 * 360 * 1e9) -> dict:
        """ Returns the row of the parameter df which has the latest timestamp 
            that is still before timestamp. If there is no timestamp to the left 
            or if the timediff is above a threshold, we return None. 
        """
        
        # Construct default return value
        default_dict = {col: np.nan for col in self.p_df.columns}
        default_dict["parameter_found"] = False
        
        # Find the best row in parameter df
        self._update_parameter_df()
        insert_index = self.p_series.searchsorted(timestamp)
        insert_index = 1 # TODO: Take away
        
        # Check if suitable row was found
        if insert_index == 0: 
            return default_dict
                
        best_row = self.p_df.iloc[[insert_index-1]]
        
        print(best_row["timestamp"].values[0])
        print(timestamp)
        
        if abs(best_row["timestamp"].values[0] - timestamp) > threshold: 
            # return default_dict
            pass
        
        # Construct return value
        dic = best_row.to_dict(orient='records')[0]
        dic["parameter_found"] = True
        
        print(dic)
        return dic
        
    def _update_parameter_df(self): 
        if self.parameter_recorder.is_up_to_date(): 
            return 
        self.p_df = self.parameter_recorder.get_table().sort_values(by=['timestamp'])
        self.p_series = pd.Series(self.p_df["timestamp"])
    
    def _load_and_clean_image_df(self): 
        df = self.image_recorder.get_table()
        df = df[df.fit_successful == True]
        return df
    
    def _load_and_clean_ssd_df(self): 
        df =  self.ssd_recorder.get_table()
        
    def _add_parameter_data(self, result_df: pd.DataFrame):
        """ Merges the parameter df into the result_df based on the timestamp. 
            Uses a binary search on the parameter rows in the background to 
            find the best match.
        """
        #self._closest_parameter_row(row["timestamp"])
        result_timestamps = result_df["timestamp"]
        matched_parameter_df = result_df.apply(lambda row: self._closest_parameter_row(row["timestamp"]), axis=1, result_type='expand')
        print(result_timestamps)
        print(matched_parameter_df)
```
